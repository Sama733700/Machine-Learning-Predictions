```python
# Calculating 95% CI and displaying histogram function

def create_confidence_interval(data):
    # Calculate the sample mean
    sample_mean = np.mean(data)

    # Calculate the standard deviation
    sample_std = np.std(data, ddof=1)

    # Calculate the standard error
    standard_error = sample_std / np.sqrt(len(data))

    # Calculate the margin of error
    margin_of_error = 1.96 * standard_error

    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    # Create a histogram of the data
    plt.figure(figsize=(8, 6))
    sns.histplot(data, kde=True)
    plt.axvline(lower_bound, color='red', linewidth=2, linestyle='--', label='Lower Bound')
    plt.axvline(upper_bound, color='green', linewidth=2, linestyle='--', label='Upper Bound')
    plt.xlabel('Data')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

    # Return the confidence interval
    return lower_bound, upper_bound
```

```python
#calculating for exam_score data
lower_bound, upper_bound = create_confidence_interval(math_exam_scores)
```
